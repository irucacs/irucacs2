Matplotlibの描画オブジェクトとサブプロット、plotメソッドによる擬似的な曲線のグラフの作成

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0.0, 10.0, 0.1)
y1 = np.sin(x)
y2 = np.cos(x)
#描画オブジェクトfigとサブプロットax
fig, ax = plt.subplots()
ax.plot(x, y1, label='sin')
ax.plot(x, y2, label='cos')
#サブプロットにタイトルを設定
ax.set_title("●axtitle")
#描画オブジェクトにタイトルを設定
fig.suptitle("★figtitle")
ax.legend()
plt.show()
```

積み上げ棒グラフの作成方法

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()

x = [1,2,3]
y1 = [9,4,2]
y2 = [2,5,8]
labels = ["apple", "banana", "melon"]
y_total =  [num1 + num2 for num1,num2 in zip(y1,y2)]
ax.bar(x,y_total, tick_label = labels, label ='y1')
ax.bar(x,y2, label = 'y2')
ax.legend()

plt.show()
```

ヒストグラムの作成方法、正規分布に従う乱数の生成

```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(123)
#平均50
mu = 50
#標準偏差10
sigma = 10 
#正規分布に従うランダム関数(平均,標準偏差,個数)
x = np.random.normal(mu, sigma, 500)
fig,ax = plt.subplots()
#binsはビン(棒グラフ)の数、orientationは向き
n, bins, patches = ax.hist(x, bins=5, orientation='horizontal')

for i, num in enumerate(n):
    print('{:.2f}-{:.2f}|{}'.format(bins[i],bins[i+1],num))

plt.show()
```

円グラフの作成方法

```python
import matplotlib.pyplot as plt
#各属性の円の中心からの距離
explode=[0, 0.2, 0]
#ラベル
labels = ['apple', 'banana', 'melon']
#各属性の値
x = [5,2,9]
fig, ax = plt.subplots()
#startangleは円グラフの開始位置、頂点は90度、指定しない場合は3時の場所から始まる。
#counterclockはFalseだと時計回り、shadowは円グラフの影、autopct引数は値の表示方法
ax.pie(x, labels = labels, startangle = 90, counterclock = False, shadow = True, autopct = '%1.2f%%', explode = explode)

plt.show()
```
